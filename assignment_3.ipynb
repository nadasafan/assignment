```python
sns.heatmap(df.corr(),annot=True)
```

above illustrate a correlation data.
from correlation coeffecient between (national income) and (per person) = 0.16 : very low correlation, 
but between (national income) and (growth) = 0.009 : very low correlation,
but between (Growth) and (per person) = -0.25 : low correlation.

```python
 plt.figure(figsize=(12, 12))
heatmap = sns.heatmap(df.corr()[['Per person']].sort_values(by='Per person', ascending=False),
                      vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Per person feature heatmap', fontdict={'fontsize':14}, pad=10)
```

```python
# sns.heatmap(df[['National Income']])
heatmap = sns.heatmap(df.corr()[['National Income']].sort_values(by='National Income', ascending=False),
                      vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('National Income feature heatmap', fontdict={'fontsize':18}, pad=16)
```

```python
heatmap = sns.heatmap(df.corr()[['Growth']].sort_values(by='Growth', ascending=False),
                      vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Growth feature heatmap', fontdict={'fontsize':18}, pad=16)
```

from above 3 figures the relation between one feature in x- axis and others' coffecient value explain as:
the coffecient value (-0.25)represent very low correlation between them.
the coffecient value (0.009)represent very low correlation between them.
the coffecient value (0.16)represent low correlation between them.
